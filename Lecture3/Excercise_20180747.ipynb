```python
# 통상적으로 첫 째 아이가 더 늦게 태어난다는 속설이 있습니다.
# 이를 확인하기 위해서 prglngth으로 첫째 아이와 첫째가 아닌 아이에 대한 평균임신기간을 계산해보세요.
# birthord에 몇번째 아이인지 나와있습니다
firstordr_prglngthMean = preg[preg.birthord==1].prglngth.mean()
notFirstordr_prglngthMean = preg[preg.birthord > 1].prglngth.mean()
print("첫째 아이 평균 임신 기간: ", firstordr_prglngthMean, "\n첫째가 아닌 아이 평균 임신 기간: ", notFirstordr_prglngthMean)
```

```python
# 첫째 아이 평균 임신 기간
preg[preg.birthord==1].prglngth.sum() / len(preg[preg.birthord==1].prglngth)
```

```python
# 첫째가 아닌 아이 평균 임신 기간
preg[preg.birthord > 1].prglngth.sum() / len(preg[preg.birthord > 1].prglngth)
```
